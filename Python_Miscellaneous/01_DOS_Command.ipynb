# DOS command execution through Python

## Approach : 1

Execute Dos command<br>Wait until Dos command execution finishes<br>Then come to next line<br>We will not be able to determine DOS command succsessfully executed or not

Package

```python
import os
```

```python
os.system("dir /S E:\\") # This DOS command will be excuted in another child process
print("DOS command execution finished")
```

```python
os.system("ABCD") # Wrong DOS command. We will never know if this command execution was successful or not
print("DOS command execution finished")
```

## Approach : 2

Execute Dos command<br>Wait until Dos command execution finishes<br>We need to know error/output from DOS command<br>Then come to next line

Package

```python
import subprocess
```

```python
DOS_Command = subprocess.Popen("dir /S E:\\", stdout=subprocess.PIPE, stderr=subprocess.PIPE, stdin=subprocess.PIPE, shell=True) # Initiate DOS Command execution
                
# Return Command output message and Error message
# Output and Error will be returned as Byte Code
# Typecast to String as needed
DOS_Command_Output, DOS_Command_Error = DOS_Command.communicate() # Script will be in waiting state until Command execution finishes
print("Command Execution finishes")
print(f"Output : {DOS_Command_Output}")
print(f"Error : {DOS_Command_Error}")
```

```python
DOS_Command = subprocess.Popen("ABCD", stdout=subprocess.PIPE, stderr=subprocess.PIPE, stdin=subprocess.PIPE, shell=True) # Initiate DOS Command execution
                
# Return Command output message and Error message
# Output and Error will be returned as Byte Code
# Typecast to String as needed
DOS_Command_Output, DOS_Command_Error = DOS_Command.communicate() # Script will be in waiting state until Command execution finishes
print("Command Execution finishes")
print(f"Output : {DOS_Command_Output}")
print(f"Error : {DOS_Command_Error}")
```

```python
subprocess.Popen("dir /S E:\\", shell=True) # DOS command starts executing in child process and immediately comeback to Parent
print("Command Execution finishes")
```
